### The problem is that a+b=c is known, and now a noise disturbance is given to a and b respectively, and the calculation formula remains unchanged, then what is the distribution of the new c relative to the original noise error? In other words, what is the relationship between its noise error distribution and the noise error distribution of a and b?

#### Take a simple example, set a=b=1, then c=2, now add a Gaussian noise to a and b respectively, and then look at the distribution of c.

```python
import random
import matplotlib
from matplotlib import pyplot as plt

A = [1]*10000
B = [1]*10000

mu,sigma = 0,0.001
C = []
for i in range(len(A)):
    A[i] += random.gauss(mu,sigma)
    B[i] += random.gauss(mu,sigma)
    C.append(A[i]+B[i])
```

```python
plt.hist(A,bins=50,color='b')
plt.show()
plt.hist(C,bins=50,color='r')
plt.show()
```

#### It can be seen that c and a still follow the Gaussian distribution, so what is the specific relationship between the two distributions? Or what is the change in the mean and variance of the two?

```python
CC = [x-1 for x in C]
plt.hist(A,bins=50,color='b',alpha=0.6)
plt.hist(CC,bins=50,color='r',alpha=0.1)
plt.show()
```

#### Here, c is processed, that is, the whole image is shifted by 1 unit to the left, so that it can be compared with the image of a.
#### It can be seen that the shapes of the image of c and the image of a are the same, but the image of c has a larger amplitude.
#### This also shows that the errors of a and b are simply superimposed, the error margin of c is just the sum of a and b, and the variances of the three are exactly the same.
